```python
import numpy as np
def read_ascii_file(filename,index):
    "This reads an ascii file of omni data"
    with open(filename) as f:
        year=[]
        day=[]
        hour=[]
        minute=[]
        data=[]
        for line in f:
            
            tmp=line.split()
            year.append(int(tmp[0])) #convert string to intergers 
            day.append(int(tmp[1]))
            hour.append(int(tmp[2]))
            minute.append(int(tmp[3]))
            data.append(int(tmp[index]))
        data=np.array(data)
        year=np.array(year)
        day=np.array(day)
        hour=np.array(hour)
        minute=np.array(minute)
        return year,day, hour,minute,data 
    
```

```python
import numpy as np
import datetime as dt
def read_ascii_file(filename,index):
    "This reads an ascii file of omni data"
    with open(filename) as f:
        #year=[]
        #day=[]
        #hour=[]
        #minute=[]
       # data=[]
        #time=[]
        data_dic={"time":[],"year":[],"day":[],"hour":[],"minute":[],"symh":[]}
        for line in f:
            tmp=line.split()
            data_dic["year"].append(int(tmp[0])) #convert string to intergers 
            data_dic["day"].append(int(tmp[1]))
            data_dic["hour"].append(int(tmp[2]))
            data_dic["minute"].append(int(tmp[3]))
            data_dic["symh"].append(int(tmp[4]))
            #create datetime in each line
            time0= dt.datetime(int(tmp[0]),1,1,int(tmp[2]),int(tmp[3]),0)\
                 + dt.timedelta(days=int(tmp[1])-1)
            data_dic["time"].append(time0)
            #data_dic["year"].append(float(tmp[index]))
        return data_dic
```

```python
filename='omni_min_case_2013.lst'
index=-1
data= read_ascii_file(filename,index)
for element in data["time"]:
    print(element.isoformat())
 
```

```python
import numpy as np
import matplotlib.pyplot as plt
data= read_ascii_file(filename,index)
time= np.array(data["time"])
data1=np.array(data["symh"])
fig,ax= plt.subplots()
y1=np.argmin(data["symh"])
y2=np.argmax(data["symh"])
plt.axvline(time[y2])
plt.axvline(time[y1])
ax.plot(time,data1,marker='.',c='gray',
       label='All Events', alpha=0.5)
ax.set_xlabel('Year of 2013')
ax.set_ylabel('SYMH (nt)')
ax.grid(True)
ax.legend
plt.show()
```
